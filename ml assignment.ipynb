### 1. What is the total sales by Customer type?

```python
df.groupby(by='Customer type').sum()
```

```python
df.groupby(by='Customer type').Total.sum()
```

### 2. Which Payment channel recorded the Highest Sales Transaction?

```python
df.groupby(by='Payment').sum()
```

```python
df.groupby(by='Payment').Total.sum()
```

```python
payment_channel = df.groupby(by='Payment').sum()[['Quantity','Unit price','new_sales_amount']]
```

```python
payment_channel.sort_values('Payment', ascending = False).iloc[0]
```

### 3. Assuming the total sales variable was not supplied, calculate the sales amount using price per unit and quantity sold as new_sales_amount.

```python
df['new_sales_amount'] = df['Unit price'] * df['Quantity']
```

```python
df.head()
```

### 4. How much goods was sold in January 2019?

```python
df[(df['Date'] >= '2019/01/01') & (df['Date'] <= '2019/01/31')]
```

### 5. Which Product is best Performing in terms of quantity sold?

```python
sales_by_product_line = df.groupby(by='Product line').sum()[['Quantity','Unit price','new_sales_amount']]
```

```python
sales_by_product_line
```

```python
sales_by_product_line.sort_values('new_sales_amount', ascending = False).iloc[0]
```

### 6. What are the Top 3 Cities with Respect to Sales Value?

```python
sales_by_city = df.groupby(by='City').sum()[['Quantity','Unit price','new_sales_amount']]
```

```python
sales_by_city
```

```python
sales_by_city.sort_values('City', ascending = False).iloc[0:3]
```
